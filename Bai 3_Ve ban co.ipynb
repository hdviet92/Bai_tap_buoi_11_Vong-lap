```python
import turtle
d = 50
tl = turtle.Turtle()
#tl.hideturtle()
tl.speed(20)

# Vẽ khung to nhất
tl.penup()
tl.goto(d*4.5, d*5)
tl.pendown()
for i in range(2):
    tl.right(90)
    tl.forward(d*10)
    tl.right(90)
    tl.forward(d*9)
    
# vẽ viền đậm   
tl.penup()
tl.goto(d*4.1, d*4.6)
tl.pensize(d/20)
tl.pendown()
for i in range(2):
    tl.right(90)
    tl.forward(d*9.2)
    tl.right(90)
    tl.forward(d*8.2)
    
# vẽ 5 hình chữ nhật ngang    
tl.pensize(1)
a = 0
while a < 5:
    tl.penup()
    tl.goto(d*4,d*(4.5 - 2*a))
    tl.pendown()
    for i in range(2):
        tl.right(90)
        tl.forward(d)
        tl.right(90)
        tl.forward(d*8)
    a += 1
    
# vẽ 9 đường dọc   
a = 0
while a < 9:
    tl.penup()
    tl.goto(d*(4-a),d*4.5)
    tl.pendown()
    tl.right(90)
    tl.forward(d*9)
    tl.left(90)
    a += 1

# vẽ 4 đường chéo trên   
a = 1
while a <= 8:
    tl.penup()
    tl.goto(0,d*3.5)
    tl.pendown()
    tl.right(45*a)
    tl.forward(d*(2**0.5))
    tl.left(45*a)
    a += 2

# vẽ 4 đường chéo dưới 
a = 1
while a <= 8:
    tl.penup()
    tl.goto(0,-d*3.5)
    tl.pendown()
    tl.right(45*a)
    tl.forward(d*(2**0.5))
    tl.left(45*a)
    a += 2
    
# vẽ biên giới 
tl.penup()
tl.goto(d*4, d/2)
tl.pendown()
tl.fillcolor('white')
tl.begin_fill()
for i in range(2):
    tl.right(90)
    tl.forward(d)
    tl.right(90)
    tl.forward(d*8)
tl.end_fill()

# Vẽ 4 hoa văn dòng ngoài
a = 0
while a <= 6:
    tl.penup()
    tl.goto((3 - a)*d, 2.5*d)
    for i in range (4):
        tl.right(45*(1+2*i))
        tl.forward(d/12)
        tl.pendown()
        tl.left(45)
        tl.forward(d/6)
        tl.backward(d/6)
        tl.right(90)
        tl.forward(d/6)
        tl.penup()
        tl.backward(d/6)
        tl.left(45)
        tl.backward(d/12)
        tl.left(45)    
    a += 6
a = 0
while a <= 6:
    tl.penup()
    tl.goto((3 - a)*d, -2.5*d)
    for i in range (4):
        tl.right(45*(1+2*i))
        tl.forward(d/12)
        tl.pendown()
        tl.left(45)
        tl.forward(d/6)
        tl.backward(d/6)
        tl.right(90)
        tl.forward(d/6)
        tl.penup()
        tl.backward(d/6)
        tl.left(45)
        tl.backward(d/12)
        tl.left(45)
    a += 6

# Vẽ 6 hoa văn trong
a = 0
while a <= 4:
    tl.penup()
    tl.goto((2 - a)*d, 1.5*d)
    for i in range (4):
        tl.right(45*(1+2*i))
        tl.forward(d/12)
        tl.pendown()
        tl.left(45)
        tl.forward(d/6)
        tl.backward(d/6)
        tl.right(90)
        tl.forward(d/6)
        tl.penup()
        tl.backward(d/6)
        tl.left(45)
        tl.backward(d/12)
        tl.left(45)
    a += 2
a = 0
while a <= 4:
    tl.penup()
    tl.goto((2 - a)*d, -1.5*d)
    for i in range (4):
        tl.right(45*(1+2*i))
        tl.forward(d/12)
        tl.pendown()
        tl.left(45)
        tl.forward(d/6)
        tl.backward(d/6)
        tl.right(90)
        tl.forward(d/6)
        tl.penup()
        tl.backward(d/6)
        tl.left(45)
        tl.backward(d/12)
        tl.left(45)
    a += 2
    
# Vẽ 4 hoa văn 1/2 
a = 0
while a <= 8:
    tl.penup()
    tl.goto((4 - a)*d, 1.5*d)
    for i in range (2):
        tl.left(45*(2+i))
        tl.forward(d/12)
        tl.pendown()
        tl.right(45)
        tl.forward(d/6)
        tl.backward(d/6)
        tl.right(90)
        tl.forward(d/6)
        tl.penup()
        tl.backward(d/6)
        tl.right(45)
        tl.backward(d/12)
        tl.right(45)
    a += 8      
        
# kết thúc vẽ
turtle.done()
```
