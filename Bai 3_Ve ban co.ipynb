```python
import turtle
d = 50
tl = turtle.Turtle()
#tl.hideturtle()
tl.speed(15)

# Vẽ khung to nhất
tl.penup()
tl.goto(d*4.5, d*5)
tl.pendown()
for i in range(2):
    tl.right(90)
    tl.forward(d*10)
    tl.right(90)
    tl.forward(d*9)
    
# vẽ viền đậm   
tl.penup()
tl.goto(d*4.1, d*4.6)
tl.pensize(d/20)
tl.pendown()
for i in range(2):
    tl.right(90)
    tl.forward(d*9.2)
    tl.right(90)
    tl.forward(d*8.2)
    
# vẽ 5 hình chữ nhật ngang    
tl.pensize(1)
a = 0
while a < 5:
    tl.penup()
    tl.goto(d*4,d*(4.5 - 2*a))
    tl.pendown()
    for i in range(2):
        tl.right(90)
        tl.forward(d)
        tl.right(90)
        tl.forward(d*8)
    a += 1
    
# vẽ 9 đường dọc   
a = 0
while a < 9:
    tl.penup()
    tl.goto(d*(4-a),d*4.5)
    tl.pendown()
    tl.right(90)
    tl.forward(d*9)
    tl.left(90)
    a += 1

# vẽ 4 đường chéo trên   
tl.penup()
tl.goto(0,d*3.5)
tl.pendown()
a = 0
while a <= 3:
    tl.right(45)
    tl.forward(d*(2**0.5))
    tl.right(135)
    tl.forward(d)
    tl.right(90)
    tl.forward(d)
    a += 1

# vẽ 4 đường chéo dưới 
tl.penup()
tl.goto(0,d*-3.5)
tl.pendown()
a = 0
while a <= 3:
    tl.right(45)
    tl.forward(d*(2**0.5))
    tl.right(135)
    tl.forward(d)
    tl.right(90)
    tl.forward(d)
    a += 1

# vẽ biên giới 
tl.penup()
tl.goto(d*4, d/2)
tl.pendown()
tl.fillcolor('white')
tl.begin_fill()
for i in range(2):
    tl.right(90)
    tl.forward(d)
    tl.right(90)
    tl.forward(d*8)
tl.end_fill()

# vẽ hoa văn trên dòng trên
a = 0
while a < 4:
    tl.penup()
    tl.goto(d*(3.9 - 2*a), d*1.4)
    tl.pendown()
    tl.right(90)
    tl.forward(d/5)
    tl.backward(d/5)
    tl.left(90)
    tl.backward(d/5)
    tl.penup()
    tl.goto(d*(3.9 - 2*a), d*1.6)
    tl.pendown()
    tl.left(90)
    tl.forward(d/5)
    tl.backward(d/5)
    tl.right(90)
    tl.backward(d/5)
    
    tl.penup()
    tl.goto(-d*(3.9 - 2*a), d*1.4)
    tl.pendown()
    tl.right(-90)
    tl.forward(-d/5)
    tl.backward(-d/5)
    tl.left(-90)
    tl.backward(-d/5)
    tl.penup()
    tl.goto(-d*(3.9 - 2*a), d*1.6)
    tl.pendown()
    tl.left(-90)
    tl.forward(-d/5)
    tl.backward(-d/5)
    tl.right(-90)
    tl.backward(-d/5)
    a += 1

# vẽ hoa văn dưới dòng trên
a = 0
while a < 4:
    tl.penup()
    tl.goto(d*(3.9 - 2*a), -d*1.6)
    tl.pendown()
    tl.right(90)
    tl.forward(d/5)
    tl.backward(d/5)
    tl.left(90)
    tl.backward(d/5)
    tl.penup()
    tl.goto(d*(3.9 - 2*a), -d*1.4)
    tl.pendown()
    tl.left(90)
    tl.forward(d/5)
    tl.backward(d/5)
    tl.right(90)
    tl.backward(d/5)
    
    tl.penup()
    tl.goto(-d*(3.9 - 2*a), -d*1.6)
    tl.pendown()
    tl.right(-90)
    tl.forward(-d/5)
    tl.backward(-d/5)
    tl.left(-90)
    tl.backward(-d/5)
    tl.penup()
    tl.goto(-d*(3.9 - 2*a), -d*1.4)
    tl.pendown()
    tl.left(-90)
    tl.forward(-d/5)
    tl.backward(-d/5)
    tl.right(-90)
    tl.backward(-d/5)
    a += 1
    
# vẽ 2 hoa văn trên dòng dưới
a = 0
while a < 2:
    tl.penup()
    tl.goto(d*(2.9 - 6*a), d*2.4)
    tl.pendown()
    tl.right(90)
    tl.forward(d/5)
    tl.backward(d/5)
    tl.left(90)
    tl.backward(d/5)
    tl.penup()
    tl.goto(d*(2.9 - 6*a), d*2.6)
    tl.pendown()
    tl.left(90)
    tl.forward(d/5)
    tl.backward(d/5)
    tl.right(90)
    tl.backward(d/5)
    
    tl.penup()
    tl.goto(-d*(2.9 - 6*a), d*2.4)
    tl.pendown()
    tl.right(-90)
    tl.forward(-d/5)
    tl.backward(-d/5)
    tl.left(-90)
    tl.backward(-d/5)
    tl.penup()
    tl.goto(-d*(2.9 - 6*a), d*2.6)
    tl.pendown()
    tl.left(-90)
    tl.forward(-d/5)
    tl.backward(-d/5)
    tl.right(-90)
    tl.backward(-d/5)
    a += 1
    
# vẽ 2 hoa văn dưới dòng dưới
a = 0
while a < 2:
    tl.penup()
    tl.goto(d*(2.9 - 6*a), -d*2.6)
    tl.pendown()
    tl.right(90)
    tl.forward(d/5)
    tl.backward(d/5)
    tl.left(90)
    tl.backward(d/5)
    tl.penup()
    tl.goto(d*(2.9 - 6*a), -d*2.4)
    tl.pendown()
    tl.left(90)
    tl.forward(d/5)
    tl.backward(d/5)
    tl.right(90)
    tl.backward(d/5)
    
    tl.penup()
    tl.goto(-d*(2.9 - 6*a), -d*2.6)
    tl.pendown()
    tl.right(-90)
    tl.forward(-d/5)
    tl.backward(-d/5)
    tl.left(-90)
    tl.backward(-d/5)
    tl.penup()
    tl.goto(-d*(2.9 - 6*a), -d*2.4)
    tl.pendown()
    tl.left(-90)
    tl.forward(-d/5)
    tl.backward(-d/5)
    tl.right(-90)
    tl.backward(-d/5)
    a += 1


# kết thúc vẽ
turtle.done()
```
